```python
def data_from_demo(test_demo,datatype='kills'):
    match_id=test_demo['matchID']
    map_name=test_demo['mapName']
    round_tot=len(test_demo['gameRounds'])
    all_kills=[]
    if not test_demo['gameRounds'] is None:
        for rnd in test_demo['gameRounds']:
            kills=rnd[datatype]
            round_num=rnd['roundNum']
            rnd_winner=rnd['ctTeam']
            if rnd['endTScore']>rnd['tScore']:
                rnd_winner=rnd['tTeam']
            if not kills is None:
                for kill in kills:
                    kill['round_won_by']=rnd_winner
                    kill['match_id']=match_id
                    kill['map_name']=map_name
                    kill['round_num']=round_num
                    kill['round_tot']=round_tot
                    all_kills.append(kill)
    print('done: ', match_id)
    return all_kills
```

```python
import glob
import os
def all_data(root_dir,datatype='kills'):
    #test=0
    downloaded=[]
    data=[]
    for filename in glob.iglob(root_dir + '**/**', recursive=True):
       # test+=1
        #if(test>10):
       #     break
        #name=os.path.basename(filename)
        if(filename.endswith('.json') and not filename in downloaded):
            print('parsing    '+filename)
            demo=load_json(filename)
            data+=data_from_demo(demo,datatype)
            downloaded.append(filename)

    with open('ALL_'+datatype+'.json', 'w') as fp:
        json.dump(data, fp)
```

```python
# build demo link to date,time,and event name dic
from bs4 import BeautifulSoup
import re

htmls=load_json('C:\\Users\\nguye\\Documents\\Python\\scraping\\major_matchpage_html.json')

```

```python
i=0
dic={}
for link in htmls:
    i+=1
    html=htmls[link]
    soup = BeautifulSoup(html)  
    print(link)
    demo_link=soup.select_one("a[href*=download]")['href']
    #div_event= html.find('div', attrs={'class': 'event text-ellipsis'})
    event_div = soup.find('div', attrs={'class': 'event text-ellipsis'})
    event=event_div.select_one("a[href*=events]")['title']
    date = soup.find('div',{'class':'date'})['data-unix']
   
    team1_div = soup.find('div', attrs={'class': 'team1-gradient'})
    team1=team1_div.find('div', attrs={'class': 'teamName'}).text
    team2_div = soup.find('div', attrs={'class': 'team2-gradient'})
    team2=team2_div.find('div', attrs={'class': 'teamName'}).text
    
    won=team2
    loss=team1
    if team1_div.find_all("div", {"class": "won"}):
        won=team1
        loss=team2
    print(i)
    dic[demo_link]=[event,date,team1,team2,won,loss]
```

```python
with open('demo_links_info.json', 'w') as fp:
        json.dump(dic, fp)
```
